```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
def get_description(imdbId):
    url = "https://www.imdb.com/title/tt"+str(imdbId)
    
    page = requests.get(url,timeout=10)
    soup = BeautifulSoup(page.content, 'html.parser')
    div = soup.find("div",class_="summary_text")
    summary = div.get_text()
    if "See full summary" in summary:
        href = div.find("a")['href']
        page = requests.get("https://www.imdb.com/"+href,timeout=10)
        soup = BeautifulSoup(page.content, 'html.parser')
        summary = soup.find("li",class_="ipl-zebra-list__item").get_text()
        summary = re.sub("[—].+[<].+[@].+[>][\n]","",summary)
        summary = re.sub("[\n][—].+[\n]","",summary)
    summary = summary.lstrip().rstrip()
    return summary
```

```python
df = pd.read_csv("links.csv", dtype={'imdbId': str})
summaries = {}
errors = []
for index, row in df.iterrows():
    movieid = row['movieId']
    imdbId = str(row['imdbId'])
    
    try:
        summary = get_description(imdbId)
        summaries[movieid] = summary

        if index%10==0:
            print(index,"/",len(df))
    except:
        errors.append((movieid,imdbId))
```

```python
fout = open("summaries.csv","a")
fout.write("movieId,summary\n")
for i in summaries:
    fout.write(str(i)+","+summaries[i]+"\n")
fout.close()
```

```python
from sklearn.model_selection import train_test_split

class Movie:
    
    def __init__(self, movieid, title, summary, genres):
        self.movieid = movieid
        self.title = title
        self.summary = summary
        self.genres = genres
        
    def to_string(self):
        return str(self.movieid)+"\t"+self.title+"\t"+self.summary+"\t"+str(self.genres)+"\n"
    
    def update_genres(self, genres_dict):
        genres = [0 for i in range(len(genres_dict))]
        for g in self.genres.split("|"):
            if g not in ["(no genres listed)","Film-Noir","IMAX"]:
                genres[genres_dict[g]] = 1
        self.genres = genres


fin = open("summaries.csv","r")
summaries = {}
for i in fin:
    i = i.rstrip()
    if "movieId" in i:
        continue
    movieid, summary = i.split("\t")
    summaries[movieid] = summary
fin.close()

fin = open("movies.csv","r")
movies = []
genres = {}
for i in fin:
    i = i.rstrip()
    if "movieId" in i:
        continue
    if "\"" in i:
        movieid, remainder = i.split(",\"")
        title, genre = remainder.split("\",")
    else:
        movieid, title, genre = i.split(",")
    if movieid not in summaries :
        continue
    movies.append(Movie(movieid, title, summaries[movieid], genre))
    genre = genre.split("|")
    for g in genre:
        if g not in genres and '(no genres listed)' not in genre and 'IMAX' not in genre and 'Film-Noir' not in genre:
            genres[g] = len(genres)
fin.close()

print(genres)
print(len(movies))

for movie in movies:
    movie.update_genres(genres)

movie_train, movie_test = train_test_split(movies, test_size=0.15, random_state=42)
fout = open("train_movies.txt","w")
for movie in movie_train:
    fout.write(movie.to_string())
fout.close()
fout = open("test_movies.txt","w")
for movie in movie_test:
    fout.write(movie.to_string())
fout.close()
```
